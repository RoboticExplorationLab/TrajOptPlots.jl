```julia
import Pkg; Pkg.activate(@__DIR__); Pkg.instantiate()
using TrajOptPlots
using TrajectoryOptimization
using RobotZoo
using StaticArrays
using Colors
using Plots
```

# Plotting 2D Obstacles

```julia
xc = [1,2,3.]
yc = [-1,-3,-2.]
rc = [0.5, 0.25, 0.75]
con = CircleConstraint(3, xc, yc, rc)
length(con)

con2 = GoalConstraint(SA[1,-2,3])
```

```julia
plot(con, circle_color=:red)
plot!(con2, xind=1, yind=2, color=:green)
plot!(con2, xind=3, yind=2, seriescolor=:yellow, markershape=:diamond, label="start")
```

# Plotting a Dubins Car
TrajOptPlots takes all of these methods together and can plot a problem for the Dubins car with a single line.

```julia
using Altro
prob, = Altro.Problems.DubinsCar(:escape);
plot(prob)
```

# RobotDynamics Methods
`RobotDynamics` also provides a few convenient methods for dealing with generic systems (which are used by the previous methods)

### Plotting states and controls

```julia
plot(states(prob), inds=1:2, labels=["x" "y"])
```

```julia
plot(controls(prob), inds=1:1, labels="linear velocity")
```

### Plotting a trajectory

```julia
using RobotDynamics
traj2(states(prob), xind=1, yind=2)
traj2!(get_trajectory(prob), lw=2, ls=:dash)
```
